```python
# Create movie_metadata classes
# ----------------------------------
class Movie(Base):
    __tablename__ = 'movie_metadata'
#     id	imdb_id	title	release_date	spoken_languages	
#     genres1	genres2	genres3	genres4	genres5	runtime	production_companies	
#     production_countries	budget	revenue	overview	tagline	vote_count	
#     homepage	popularity	actors	boxoffice	director	poster_link	
#     production_company	vote_average	vote_average_perc	
#     Internet_Movie_Database_rating	Rotten_Tomatoes_rating	
#     Metacritic_rating	rating_average	writer
    
    id = Column(Integer, primary_key=True)
    imdb_id = Column(String(255))
    title = Column(String(255))
    release_date = Column(String(255))
    spoken_languages = Column(String(255))
    genres1 = Column(String(255))
    genres2 = Column(String(255))
    genres3 = Column(String(255))
    genres4 = Column(String(255))
    genres5 = Column(String(255))
    runtime = Column(String(255))
    production_companies = Column(String(255))
    production_countries = Column(String(255))
    budget = Column(String(255))
    revenue = Column(String(255))
    overview = Column(String(255))
    tagline = Column(String(255))
    vote_count = Column(Float)
    homepage = Column(String(255))
    vote_average = Column(Float)
    vote_average_perc = Column(Float)
    Internet_Movie_Database_rating = Column(Float)
    Rotten_Tomatoes_rating = Column(Float)
    Metacritic_rating = Column(Float)
    rating_average = Column(Float)
    
    
    popularity = Column(String(255))
    actors = Column(String(255))
    boxoffice = Column(String(255))
    director = Column(String(255))
    poster_link = Column(String(255))
    production_company = Column(String(255))
    ratings = Column(String(255))
    writer = Column(String(255))
    
   
```

```python
# Create Database Connection
# ----------------------------------
# Establish Connection
engine = create_engine("sqlite:///movieData.sqlite")
conn = engine.connect()
```

```python
# Create table within the database
Base.metadata.create_all(conn)
```
